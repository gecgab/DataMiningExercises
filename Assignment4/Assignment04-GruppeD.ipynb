### 3.2a)

```python
import matplotlib.pyplot as plt

def calcP(d):
    P = []
    for elem in d:
        P.append(elem/sum(d))   
    return P

p = [1,3,4,6,10]

d1 = (p[0]-3)**2
d2 = (p[1]-3)**2
d3 = (p[2]-3)**2
d4 = (p[3]-3)**2
d5 = (p[4]-3)**2

d = [d1, d2,d3,d4, d5]

print("d1: ",d1) 
print("d2: ",d2)
print("d3: ",d3)
print("d4: ",d4)
print("d5: ",d5)
```

### 3.2b)

```python
P = calcP(d)
plt.plot(d,P,'bo')

print("P1: ",P[0],"P2: ", P[1],"P3: ", P[2],"P4: ", P[3],"P5: ", P[4]) 
plt.show()
```

## 3.2c)

```python
def calcDis(p,c):
    minDis = (p-c[0])**2
    minC = 0
    for i,ci in enumerate(c):
        d = (p-ci)**2
        if d < minDis:
            minDis = d 
            minC = i
    return minDis

c = [3,10]

d = []
for pi in p:
    d.append(calcDis(pi,c))
    
print("min d1: ",d[0])
print("min d2: ",d[1])
print("min d3: ",d[2])
print("min d4: ",d[3])
print("min d5: ",d[4])

Pnew = calcP(d)

print("P1: ",Pnew[0])
print("P2: ",Pnew[1])
print("P3: ",Pnew[2])
print("P4: ",Pnew[3])
print("P5: ",Pnew[4])
```
