```python
import numpy as np
import pickle
import matplotlib.pyplot as plt # for plotting images
from sklearn.cluster import KMeans

data, labels = pickle.load(open('digits.pickle', 'rb'))

k = len(np.unique(labels)) # should be 10

kmeans = KMeans(n_clusters=k, random_state=1337, n_init=1).fit(data) # n_init=1 since we use deterministic seed
centroids = kmeans.cluster_centers_

fig, axs = plt.subplots(2,5)
fig.subplots_adjust(top=1,right=1.5) # enlarge individual plots
for i in range(k):
    centroid = centroids[i,:].reshape(16,16)
    axs[int(i/5),i%5].imshow(centroid, cmap="gray")
    axs[int(i/5),i%5].axis('off')
plt.show()
```
